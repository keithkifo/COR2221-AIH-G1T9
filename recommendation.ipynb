```python
import requests
import os
from dotenv import load_dotenv
from pprint import pprint
load_dotenv()

STB_API_KEY = os.getenv('STB_API_KEY')
```

```python
import random
food_category = random.choice( ['dim sum', 'korean', 'hawker food', 'pizza',' western', 'malay', 'chinese'] )
food_category


# ==== Call STB API ====
endpoint = 'https://api.stb.gov.sg/content/common/v2/search'

req = requests.get(
    endpoint,
    headers={'X-API-Key': STB_API_KEY},
    params={
        'dataset': 'food_beverages,bars_clubs',
        'keyword': food_category,
    }
)

resp = req.json()['data']

# ==== Choose a Recommendation ====
print(resp)
recommendation = random.choice( resp )
print(recommendation)

response_dict = {
    "name": recommendation['name'],
    "type": recommendation['type'],
    "description": recommendation['description'],
    "tags": [ '#' + tag.replace(' ', '_') + ' ' for tag in recommendation['tags'] ],
    "website": recommendation['officialWebsite'],
    "image_uuid": recommendation['thumbnails'][0]['uuid'] if len( recommendation['thumbnails'] ) != 0 else False
}
```

## API: Search Multiple Datasets by Keyword

```python
endpoint = 'https://api.stb.gov.sg/content/common/v2/search'

req = requests.get(
    endpoint,
    headers={'X-API-Key': STB_API_KEY},
    params={
        'dataset': 'walking_trails,tours,events,venues',
        'keyword':'museum',
        'startDate': '2023-03-31'
    }
)

resp = req.json()
```

```python
def get_media( uuid ):
    media_endpoint = f'https://api.stb.gov.sg/media/download/v2/{uuid}'

    req = requests.get(
        media_endpoint,
        headers={'X-API-Key': STB_API_KEY},
        params={
            'fileType': 'Default',
        }
    )
    return req
```
